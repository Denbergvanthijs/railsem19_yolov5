Via a Python script ([credits](https://github.com/ultralytics/yolov5#quick-start-examples)):

```python
import matplotlib.pyplot as plt
import torch

%matplotlib inline

model = torch.hub.load("ultralytics/yolov5", "custom", "./models/trained_best.pt")
model.classes = [0, ]  # Only enable relevant classes

# Source
path = "./data/rs19_val/jpgs/rs19_val/rs00106.jpg"

# Inference
results = model(path)

# Results
results.print()  # or .show(), .save(), .crop(), .pandas(), etc.
plt.imshow(results.imgs[0])
plt.show()
```
